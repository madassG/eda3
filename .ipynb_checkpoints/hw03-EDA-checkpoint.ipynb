### Задание 2 (0.5 балла)

Определите, есть ли пропуски в данных. Разделите признаки на количественные, бинарные, порядковые и категориальные.

```python
# Количественные
# 1.people_liable
# 2.duration
# 3.amount
# 4.age
# 5.number_credits

# Бинарные
# 1.credit_risk
# 2.foreign_worker
# 3.telephone

# Порядковые
# 1.status (мб категориальный) 2.savings (мб категориальный)

# Категориальные
# 1.credit_history
# 2.purpose
# 3.employment_duration (мб порядковый)
# 4.installment_rate
# 5.personal_status_sex
# 6.other_debtors
# 7.present_residence (мб порядковый)
# 8.property
# 9.other_installment_plans
# 10.housing
# 11.job 

# 'status', 'duration', 'credit_history', 'purpose', 'amount', 'savings',
# 'employment_duration', 'installment_rate', 'personal_status_sex',
# 'other_debtors', 'present_residence', 'property', 'age',
# 'other_installment_plans', 'housing', 'number_credits', 'job',
# 'people_liable', 'telephone', 'foreign_worker', 'credit_risk'

data.isnull().sum()
# No nulls

data['status_dm_l0'] = data['status'] == '... < 0 DM'
data['status_dm_ge1_l200'] = data['status'] == '0<= ... < 200 DM'
data['status_dm_ge200'] = data['status'] == '... >= 200 DM / salary for at least 1 year'
data['status_dm_noacc'] = data['status'] == 'no checking account'
# data.drop('status')

data['duration'].describe()

data['credit_risk'].unique()
data['credit_risk_bool'] = data['credit_risk'] == 'good'
# data.drop('credit_risk')

data['foreign_worker_bool'] = data['foreign_worker'] == 'yes'
# data.drop('foreign_worker')

data['telephone'].unique()
data['telephone_bool'] = (data['telephone'] != 'no')
# data.drop('telephone')

data['people_liable'].unique()
data['people_liable_ge3'] = data['people_liable'] == '3 or more'
# data.drop('people_liable')

data['job_unemloyed_n_resident'] = data['job'] == 'unemployed/unskilled - non-resident'
data['job_unemployed_resident'] = data['job'] == 'unskilled - resident'
data['job_skilled_official'] = data['job'] == 'skilled employee/official'
data['job_self_emplyed_qualified'] = data['job'] == 'manager/self-empl./highly qualif. employee'
# data.drop('job')

data['housing'].unique()
data['housing_for_free'] = data['housing'] == 'for free'
data['housing_rent'] = data['housing'] == 'rent'
data['housing_own'] = data['housing'] == 'own'
# data.drop('housing')

data['other_installment_plans'].unique()
data['other_installment_plans_none'] = data['other_installment_plans'] == 'none'
data['other_installment_plans_bank'] = data['other_installment_plans'] == 'bank'
data['other_installment_plans_stores'] = data['other_installment_plans'] == 'stores'
# data.drop('other_installment_plans')

data['age'].value_counts()
labels = [1,2,3,4, 5, 6]
bins = [18, 20, 30, 40, 50, 60, 100]
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, right=True)

data['property'].unique()
data['property_car_or_other'] = data['property'] == 'car or other'
data['property_unknown'] = data['property'] == 'unknown / no property'
data['property_savings'] = data['property'] == 'building soc. savings agr./life insurance'
data['property_estate'] = data['property'] == 'real estate'
# data.drop('property')

data['present_residence'].unique()
# ['>= 7 yrs', '1 <= ... < 4 yrs', '4 <= ... < 7 yrs', '< 1 yr']
data['present_residence_ordinal'] = 1
data.loc[data['present_residence'] == '1 <= ... < 4 yrs', 'present_residence_ordinal'] = 2
data.loc[data['present_residence'] == '4 <= ... < 7 yrs', 'present_residence_ordinal'] = 3
data.loc[data['present_residence'] == '>= 7 yrs', 'present_residence_ordinal'] = 4
# data.drop('present_residence')

data['other_debtors'].unique()
data['other_debtors_none'] = data['other_debtors'] == 'none'
data['other_debtors_guarantor'] = data['other_debtors'] == 'guarantor'
data['other_debtors_coapplicant'] = data['other_debtors'] == 'co-applicant'
# data.drop('other_debtors')

data['personal_status_sex'].unique()
personal_status_sex_factor = pd.get_dummies(data['personal_status_sex'], prefix='sex_status')
data = pd.concat([data, personal_status_sex_factor], axis=1)
# data.drop('personal_status_sex')

data['installment_rate'].unique()
installment_rate_factor = pd.get_dummies(data['installment_rate'], prefix='installment_rate')
data = pd.concat([data, installment_rate_factor], axis=1)
# data.drop('installment_rate')

data['employment_duration'].unique()
employment_duration_factor = pd.get_dummies(data['employment_duration'], prefix='employment_duration')
data = pd.concat([data, installment_rate_factor], axis=1)
# data.drop('employment_duration')

data['savings'].unique()
savings_factor = pd.get_dummies(data['savings'], prefix='savings_factor')
data = pd.concat([data, savings_factor], axis=1)
# data.drop('savings')

data['purpose'].unique()
purpose_factor = pd.get_dummies(data['purpose'], prefix='purpose_factor_')
data = pd.concat([data, purpose_factor], axis=1)
# data.drop('purpose')

data['credit_history'].unique()
credit_history_factor = pd.get_dummies(data['credit_history'], prefix='credit_history_factor_')
data = pd.concat([data, credit_history_factor], axis=1)
# data.drop('credit_history')

data['number_credits'].unique()
number_credits_factor = pd.get_dummies(data['number_credits'], prefix='number_credits_factor_')
data = pd.concat([data, number_credits_factor], axis=1)
# data.drop('number_credits')

data['duration'].unique()

# preparation done! All factor features are one-hot coded
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Используя numpy и matplotlib, начнём с тренировки: постройте график функций $sin(x)$ и $cos(x)$ от $-\pi$ до $\pi$ из 100 точек c легендой и сеткой. Используйте для линий разные цвета, одна из них пусть будет пунктирной. Результат в этом и следующих заданиях на построение графиков должен получиться примерно как на картинке. Содержание подписей не должно обязательно совпадать, однако осмысленные подписи должны присутствовать на графиках.

```python
x = np.linspace(-np.pi, np.pi, 100)

y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()

ax.grid(True)

ax.plot(x, y1, label = 'sin(x)', color='cyan', linewidth=2)
ax.plot(x, y2, '--', label = 'cos(x)', color='purple', linewidth=2)

ax.legend()

fig.set_figheight(5)
fig.set_figwidth(8)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
X = df.values
y = data['credit_risk'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
neigh = KNeighborsClassifier(n_neighbors=25)
neigh.fit(X_train, y_train)

predicted = neigh.predict(X_test)

accuracy_score(y_test, predicted)
```
